```python
import pandas as pd
```

```python
train_ratio = 0.9
train_file_path = './train_curated.csv'
output_train_file_path = './train_split.csv'
output_eval_file_path = './eval_split.csv'
```

```python
train_labels = pd.read_csv(train_file_path)
train_labels['categories'] = pd.Categorical(train_labels['labels'])
train_labels['categories'] = train_labels['categories'].cat.codes
```

```python
train_labels
```

```python
labels_per_cat = [g[1] for g in train_labels.groupby('categories')]
train_files_per_category = [files[:round(train_ratio * len(files))] for files in labels_per_cat]
eval_files_per_category = [files[round(train_ratio * len(files)):] for files in labels_per_cat]
```

```python
train_files_per_category
```

```python
train_df = []
for category in train_files_per_category:
    train_df += [pd.DataFrame(category)]
    
train_df = pd.concat(train_df)

train_df.drop(labels=['categories'], axis=1, inplace=True)

train_df
```

```python
eval_df = []
for category in eval_files_per_category:
    eval_df += [pd.DataFrame(category)]
    
eval_df = pd.concat(eval_df)

eval_df.drop(labels=['categories'], axis=1, inplace=True)

eval_df
```

```python
train_df.to_csv(output_train_file_path, index=False)
eval_df.to_csv(output_eval_file_path, index=False)
```
